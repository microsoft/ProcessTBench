```python
from utils import data_multimedia
import matplotlib.pyplot as plt
import networkx as nx
from utils import create_graph
```

```python
fig, axes = plt.subplots(2, 2, figsize=(30, 20), dpi=300)  # Adjust to 4x5 grid and increase figure size
axes = axes.flatten()

for i in range(4):  # Adjust range to 20
    G = create_graph(data_multimedia[i+7])
    pos = nx.planar_layout(G)
    nx.draw_networkx_nodes(G, pos, node_color="lightgray", node_size=1200, ax=axes[i])
    nx.draw_networkx_edges(G, pos, arrows=True, arrowsize=30, ax=axes[i])
    nx.draw_networkx_labels(G, pos, font_size=18, font_color="black", font_weight="bold", ax=axes[i])
    
    title = G.graph['user_request']
    max_width = 50
    title = '\n'.join([title[i:i+max_width] for i in range(0, len(title), max_width)])
    
    axes[i].set_title(title, fontsize=22)
    axes[i].axis("off")

plt.tight_layout(pad=7.0)
plt.show()
```
